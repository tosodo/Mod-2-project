```python
import sqlite3
import pandas as pd
import matplotlib.pyplot as plt
import  requests 

```

```python
# Creating a variable to access data 
conn = sqlite3.connect('database.sqlite')
cursor = conn.cursor()
```

```python
#c is the variable equals to cursor.execute
c = cursor.execute("""
            SELECT * FROM matches
            WHERE Season == 2011
            AND Div != 'E0'
            """)

df = pd.DataFrame(c.fetchall())
df.columns = [x[0] for x in c.description]
df.head()
```

```python
#create a dummy variable (dummy)
dummy = pd.get_dummies(df['FTR'])
dummy.columns = ['Loss','Draw','Win']
dummy.head()
```

```python
#concatenate the data with pd concat method  
df_homes = pd.concat([df,dummy], axis=1,join='inner')
df_homes.head()
```

```python
dfh = df_homes.groupby('HomeTeam').sum()
```

```python
dfa = df_homes.groupby('AwayTeam').sum()
```

```python
dfnew = dfa
dfnew['Total_Goals'] =  dfh['FTHG'] + dfa['FTAG']
dfnew.head()
```

```python
dfnew['Total_Goals'] = dfa['FTAG'] + dfh['FTR']
dfnew.head()
```

```python
dfnew['Total_Draws'] = dfa['FTAG'] + dfh['FTHG']
dfnew.head()
```

```python
dfnew['Total_Loss'] = dfa['Wins'] + dfh['Draws']
dfnew.head()
```
